```python
#Group BY
```

```python
## the groupby() function in pandas is used to group data based on one or more columns in DataFrames
```

```python
df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
S_Qty=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=S_Qty)
```

```python
## Note:From above Graph you can see that most of the Qty buys M-Size in the sales
```

```python
#Histogram
df['Size'].hist()
```

```python
df['Category']=df['Category'].astype(str)
column_data=df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=30,edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

```python
##Note:Frome above Graph you can see that most of the buyers are T-shirt
```

```python
#Checking B2B Data by using pie chart
B2B_Check=df['B2B'].value_counts()

#plot the pie chart
plt.pie(B2B_Check,labels=B2B_Check,autopct='%1.1f%%')
plt.show()
```

```python
#Checking B2B Data by using pie chart
B2B_Check=df['B2B'].value_counts()

#plot the pie chart
plt.pie(B2B_Check,labels=B2B_Check.index,autopct='%1.1f%%')
plt.show()
```

```python
## Note: From aove chart we can see that maximum i.e. 99.3% of buyers are retailers and 0.8% are B2B buyers
```

```python
#preoare data for scatter plot
x_data=df['Category']
y_data=df['Size']

#plot scatter data
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
#Plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of States')
plt.xticks(rotation=90)
plt.show()
```

```python
#top_10 States
top_10_States=df['ship-state'].value_counts().head(10)
#Plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_States.index)],x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of States')
plt.xticks(rotation=45)
plt.show()
```

```python
 ## Note:From above Graph you can see that most of the buyers are Maharashtra state
```
